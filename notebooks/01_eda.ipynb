```python
import pandas as pd

df = pd.read_excel("../data/Cleaned_Road_Safety_Data_Fixed.xlsx")

df.head()
```

```python
print(df.columns)
```

```python
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.lineplot(data=df, x='Year', y='Road Deaths', marker='o')

plt.title('Road Death Trend in Malaysia (2000–2016)')
plt.xlabel('Year')
plt.ylabel('Road Deaths')

plt.grid(True)

plt.show()

# main chart use in report , slides , StreamLit app
```

```python
vehicle_cols = [
    'Bicycle', 'Bus', 'Jeep', 'Motorcar',
    'Motorcycle', 'Others', 'Pedestrian',
    'Trailer/Lorry', 'Vans'
]

vehicle_totals = df[vehicle_cols].sum()

plt.figure(figsize=(12,6))

sns.barplot(x=vehicle_totals.index, y=vehicle_totals.values)

plt.title('Total Fatalities by Road User Type')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')        

plt.show()
```
